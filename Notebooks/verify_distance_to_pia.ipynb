```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
def plot(project_path, project_prefix):
            
    pixel_size = 0.3460130331522824
    
    with open(project_path + 'SLD_0000' + project_prefix + '_out_of_pia.pkl', 'rb') as fd:
        pia = pickle.load(fd)
    with open(project_path + 'SLD_0000' + project_prefix + '_points_annotations.pkl', 'rb') as fd:
        points = pickle.load(fd)
    with open(project_path + 'SLD_0000' + project_prefix + '_S1HL_annotations.pkl', 'rb') as fd:
        sh1l = pickle.load(fd)
    with open(project_path + 'SLD_0000' + project_prefix + '_cells_features.pkl', 'rb') as fd:
        features = pickle.load(fd)
        

    points_x = points['Centroid X µm']
    points_y = points['Centroid Y µm'] 
    plt.scatter(points_x, points_y, label = 'points', c ='b', s=100)

    sh1l_x = sh1l['Centroid X µm']
    sh1l_y = sh1l['Centroid Y µm']
    plt.scatter(sh1l_x, sh1l_y, label = 'sh1l', c ='r')



    pia_x = pia['Centroid X µm'] 
    pia_y = pia['Centroid Y µm']

    plt.scatter(pia_x, pia_y, label = 'pia', c ='g')



    feat_x = features['Centroid X µm'] 
    feat_y = features['Centroid Y µm']
    print(feat_x)
    distance = features['Distance to annotation with Outside Pia µm'] +   points_y.min()
    plt.scatter(feat_x, feat_y,s=0.01, c='y', label='cells')
    

    plt.scatter(feat_x, distance, label = 'distance_to_pia', s=0.1, c ='cyan')
    
    plt.legend()
    plt.title(project_prefix)

    plt.gca().invert_yaxis()
    plt.show()
```

```python
#project_path = './datasets/QPath_exported_Pandas_Dataframe/Dataframe/Nissl_2/01413829_RH_Nissl_2/'
project_path = './results/01413829_RH/15_12_23/'
plot(project_path, '485.vsi - 20x_03')
```
