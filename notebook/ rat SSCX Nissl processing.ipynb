```python
import numpy as np
import matplotlib.pyplot as plt
```

# Create fake data

```python
button=5.0
top=20.0
nb_slice = 10
```

```python
#data = np.random.poisson(lam=(100., 100.), size=(100, 2))
```

```python
data = (np.random.random((1000,3))*(top-button)+button) 
```

```python
x = data[:,0]
y = data[:,1]
z = data[:,2]
```

# compute fractions

```python
length = top - button
fractions=np.zeros(data.shape[0], dtype=float)
for index, point in enumerate(data):
    fractions[index]=(point[1]-button) / (length)
```

# Plot poistions and fraction histogram

```python
plt.subplot(121)
_ = plt.scatter(x, y, s=1)
plt.subplot(122)
fract_hist = plt.hist(fractions, nb_slice, orientation='horizontal')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
```

```python
nb_points_per_bin = fract_hist[0]
print(nb_points_per_bin)
bins = fract_hist[1]
print(bins)
```

# Compute densities from fractions and volume

```python
slices_x_length = np.random.random(nb_slice) * 20
slices_y_length = np.random.random(nb_slice) * 20
slices_z_length = np.random.random(nb_slice) * 20
```

```python
densities = nb_points_per_bin / (slices_z_length * slices_x_length * slices_y_length)
```
